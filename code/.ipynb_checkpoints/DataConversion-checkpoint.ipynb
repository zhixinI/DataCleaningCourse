### 数据转换

#### 1. 日期格式数据处理

```python
import numpy as np
import pandas as pd
```

```python
#encoding:指定字符集类型;dtype:每列数据的数据类型。
df = pd.read_csv('../data/baby_trade_history.csv',encoding='utf-8',dtype={'user_id':str})
```

```python
df.head()
```

```python
#对购买日期进行转换;加errors防止报错
df['buy_date'] = pd.to_datetime(df['day'],format='%Y%m%d',errors='coerce')
```

```python
df
```

```python
#查看类型
df.dtypes
```

```python
#可以提取对应年月日，必须为timestamp才可以，具体的时间点
df['buy_date'].dt.year
```

#### 2. 字符串数据处理

- 摩托车的销售情况

- Condition：摩托车新旧情况（new：新的 和used：使用过的）
- Condition_Desc：对当前状况的描述
- Price：价格
- Location：发获地址
- Model_Year：购买年份
- Mileage：里程
- Exterior_Color：车的颜色
- Make：制造商（牌子）
- Warranty：保修
- Model：类型
- Sub_Model：车辆类型
- Type：种类
- Vehicle_Title：车辆主题
- OBO：车辆仪表盘
- Watch_Count：表数

```python
df2 = pd.read_csv('../data/MotorcycleData.csv',encoding='gbk')
df2.head()
```

```python
#字符串切片
df2['Price'].str[0:4]
```

```python
#把Price的object转换为float
df2['Price'].astype(float)
```

```python
#直接转换会报错，需要进行字符串处理
df2['价格'] = df2['Price'].str.strip('$')
df2['价格'] = df2['价格'].str.replace(',','')
df2['价格'] = df2['价格'].astype(float)
```

```python
df2.Model.head(5)
```

```python
#对文本的处理，使用字符串分割
df2['Location'].str.split(',')
```

```python
#计算字符串长度
df2['Mileage'].str.len()
```

3. 高阶函数数据处理

```python
#婴儿信息
df3 = pd.read_csv('../data/sam_tianchi_mum_baby.csv',encoding='utf-8',dtype=str)
```

```python
df3
```

```python
#自定义函数
def f(x):
    if '0' in x:
        return '女'
    elif '1' in x:
        return '男'
    else:
        return '未知'
```

```python
df3['性别'] = df3['gender'].apply(f)
df3.head(3)
```

```python
#使用map函数
df3['性别4'] = df3['gender'].map({'0':'女','1':'男','2':'未知'})
df3
```

```python
#结合lambda使用
df3['user_id'].apply(lambda x : x.replace(x[1:3],'**'))
```

```python
#提取年份
df3['birthday'].apply(lambda x : str(x)[0:4])
```
